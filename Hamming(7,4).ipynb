```python
def intToByte(n, padding=8):
    byte = []
    while n > 0:
        byte.append(int(n%2))
        n = int((n - n%2)/2)
    byte = byte + [0]*(padding-len(byte))
    return byte

def parity(arr):
    return sum(arr)%2

def encoding(s):
    return s + [parity([s[0],s[1],s[2]]), parity([s[1],s[2],s[3]]), parity([s[0],s[2],s[3]])]

def decoding(r):
    disks = [parity([r[0],r[1],r[2],r[4]]), parity([r[1],r[2],r[3],r[5]]), parity([r[0],r[2],r[3],r[6]])]
    if disks[0] and disks[2] and not disks[1]:
        r[0] = 1- r[0]
    elif disks[0] and disks[1] and not disks[2]:
        r[1] = 1-r[1]
    elif disks[1] and disks[2] and not disks[0]:
        r[3] = 1-r[3]
    elif disks[0] and disks[1] and disks[2]:
        r[2] = 1-r[2]
    return r

for i in range(0,16):
    c1,c2,c3,c4 = 0,0,0,0
    s = intToByte(i,4)
    for j in range(0,127):
        err = intToByte(j,7)
        f = sum(err)
        prob = pow(0.1,f)*pow(0.9,8-f)
        r = decoding(list(map(parity, zip(encoding(s),err))))[:4]
        if r[0] != s[0]:
            c1 += prob
        if r[1] != s[1]:
            c2 += prob
        if r[2] != s[2]:
            c3 += prob
        if r[3] != s[3]:
            c4 += prob
    print(c1,c2,c3,c4)
```
